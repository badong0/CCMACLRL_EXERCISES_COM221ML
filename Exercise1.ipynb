# LIST COMPREHENSION #

```python
newlist = [x for x in fruits]
print(newlist)
```

```python
mynewlist = []
for i in fruits:
  mynewlist.append(i)

mynewlist
```

```python
newlist1 = [x for x in fruits if "p" in x]
newlist1

newlist2 = [x for x in fruits if "a" in x]
newlist2
```
